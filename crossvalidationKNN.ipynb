```python
def openFile():
  DataTrain = []
  try:
    with open('student-por_datatrain.csv') as filecsv:
      dataa = csv.reader(filecsv)
      for row in dataa:
        try:
          n = [row[0],row[1],row[2],row[3],row[4],row[5],row[6],row[7],row[8],row[9],row[10],row[11],row[12],row[13],row[14],row[15],row[16],row[17],
               row[18],row[19],row[20],row[21],row[22],row[23],row[24],row[25],row[26],row[27],row[28],row[29],int(row[30]),int(row[31]),int(row[32]),int(row[33])]
          DataTrain.append(n)
        except ValueError as error:
            DataTrain.append(row)
            continue
      DataTrain.pop(0)
      return DataTrain
  except EnvironmentError as error:
    print ("Error")
```

```python
def filejarak(filelist, fold):
  foldedList = [None] * fold
  for i in range(1, fold + 1):
    temp = []
    bawah = ((549 // fold) * i)
    atas = ((549 // fold) * (i - 1)) + 1
    for j in range(atas - 1, bawah):
      temp.append(filelist[j])
      foldedList[i - 1] = temp 
  return foldedList
```

```python
def cariJarak(hasil,listtjarak):
    distances = []
    for t in range (0,listtjarak[hasil].__len__()) :
        distances.append([])
        for i in range (0,listtjarak.__len__()) :
            if i != hasil:
                for j in range (0,listtjarak[i].__len__()) :
                    x1 =listtjarak[hasil][t][30]
                    x2 =listtjarak[hasil][t][31]
                    x3 = listtjarak[hasil][t][32]
                    y1 = listtjarak[i][j][30]
                    y2 = listtjarak[i][j][31]
                    y3 = listtjarak[i][j][32]
                    Mjob =x1-y1
                    G1 =x2-y2
                    G2 =x3-y3
                    hitung=((math.sqrt((Mjob**2)+(G1**2)+(G2**2)),i,j))
                    distances[t].append(hitung)
    for i in range (0,distances.__len__()):
        distances[i].sort()
    return distances
```

```python
def Kelas(jarak, filejarak1, KNN):
    ya,tidak= 0,0
    for i in range(0, KNN):
        indeksp,indeksc = jarak[i][1],jarak[i][-1]
        if (filejarak1[indeksp][indeksc][33] == 1):
            ya+=1
        else:
            tidak+= 1
    if tidak>ya:
        return 0
    else:
        return 1
```

```python
def crossvalidation(fold, filejarak1, KNN):
    hasil = 0
    akurasi,SortedTestData = [],[]
    while hasil < fold:
        jk = cariJarak(hasil,filejarak1)
        SortedTestData.append(jk)
        hasil+=1
    for i in range(0, SortedTestData.__len__()):
        sama = 0
        for j in range(0, SortedTestData[i].__len__()):
            n1 = Kelas(SortedTestData[i][j],filejarak1,KNN)
            n2 = filejarak1[i][j][33]
            if (n1 == n2):
                sama+=1
        akr = sama/SortedTestData[i].__len__()
        akurasi.append(akr)
    akurasitotal=sum(akurasi)/akurasi.__len__()
    return akurasitotal
```

```python
def hitung(KNN):
    file = openFile()
    fold = 10
    filejarakk = filejarak(file, fold)
    return crossvalidation(fold, filejarakk, KNN)
```

```python
def Main():
    Knnterbaik, Akurasiterbaik = 0,0
    arrAkurasi = []
    for knn in range(1,549):
      Akurasi = hitung(knn)*100
      print (knn, ".", Akurasi)
      arrAkurasi.append([knn,Akurasi])
      if Akurasi > Akurasiterbaik:
        Akurasiterbaik = Akurasi
        Knnterbaik = knn
    return (Knnterbaik)
Main()
# with open('Result.csv','HasilTes') as filecsv:
#     writer = csv.writer(filecsv)
#     writer.writerows(arrAkurasi)
```
