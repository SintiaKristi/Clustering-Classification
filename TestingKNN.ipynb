```python
#menghitung jarak pakai euclied
#menghitung list, contoh: hasil[1], berarti like
#untuk mencari k yang terbaik
def HitJarak(hasil, listjarak):
    jarak = []
    for i in range(0, 549):
      x1 = hasil[1]
      x2 = hasil[2]
      x3 = hasil[3]
      y1 = listjarak[i][1]
      y2 = listjarak[i][2]
        y3 = listjarak[i][3]
        Mjob = x1-y1
        G1 = x2-y2
        G2 = x3-y3
        hitung = Mjob ** 2 + G1 ** 2 + G2 ** 2 
        jarak.append((math.sqrt(hitung), i))
    return jarak
```

```python
#untuk mengklaskan termasuk hoaks apa enggak
#hasil jaraknya diurutkan(distance)
def Kelas(distance, DataTrain, KNN):
    distance.sort()
    ya,tidak = 0,0
    for i in range(0, KNN):
        print (DataTrain[distance[i][1]][4]) #5 dari like provokasikomentar emosi hoaks
        if DataTrain[distance[i][1]][4] == "1":
            ya += 1
        else:
            tidak += 1
    print (ya,tidak) #untuk membandingkan dari hasil yang sudah dihasilkan ya dan tidak
    if  (ya>tidak):
        return 1
    else:
        return 0
```

```python
#baca file
def openFile(path):
    temp = []
    try:
        with open(path) as csvFile:
            data = csv.reader(csvFile)
            #dijadikan bentuk list
            for row in data:
                try:
                    #convert data jadi integer dari like, provokasi, dll
                    temp.append([row[0],int(row[1]),int(row[2]),int(row[3]),int(row[4])])
                    #try cath apabila error ngasih notifikasi
                except ValueError as eror:
                    temp.append(row)
                    continue
            temp.pop(0)
            return temp
    except EnvironmentError as eror:
        print (" error")
```

```python
def main(KNN):
    #open file dalam bentuk csv untuk data train dan data tes
    DataTrain = openFile('student-por_datatrain2.csv')
    DataTesting = openFile('student-por_datatest2.csv')
    jarak = []
    #menghitung jarak dari datasetnya
    for i in range(0, DataTesting.__len__()):
        jarak.append(HitJarak(DataTesting[i], DataTrain))
    pilih = []
    #memilih kelas dari hasil perhitungan jarak
    for i in range(0, jarak.__len__()):
        pilih.append(Kelas(jarak[i], DataTrain, KNN))
    print(pilih)
# k yang paling bagus adalah 69
main(4)
```
